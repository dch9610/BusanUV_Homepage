```python
import pandas as pd
import csv
```

```python
busan_fun = pd.read_csv('./소상공인시장진흥공단_상가(상권)정보_부산_202009.csv', encoding='UTF-8', sep='|')
```

```python
busan_fun
```

```python
busan_fun = busan_fun[['상호명','상권업종대분류명','상권업종중분류명','상권업종소분류코드','상권업종소분류명','도로명주소','경도','위도']]
busan_fun
busan_fun.rename({"상호명":'name',
                  '상권업종대분류명':'clasificar',
                  '상권업종중분류명':'middle clasificar',
                  '상권업종소분류코드':'code',
                  '상권업종소분류명':'small clasificar',
                  '도로명주소':"addr",
                  '경도':'long',
                  '위도':'lat'}, axis = 'columns' ,inplace= True)
```

```python
busan_fun
```

```python
busan_entertainment=busan_fun.loc[busan_fun["middle clasificar"].str.contains("PC/오락/당구/볼링등", na=False)]
```

```python
busan_entertainment['addr'].unique()
```

```python
data1=busan_entertainment.loc[busan_entertainment["addr"].str.contains("부산대학로", na=False)]
data2=busan_entertainment.loc[busan_entertainment["addr"].str.contains("금정로", na=False)]
data3=busan_entertainment.loc[busan_entertainment["addr"].str.contains("금강로", na=False)]
data4=busan_entertainment.loc[busan_entertainment["addr"].str.contains("장전로", na=False)]
```

```python
result = pd.concat([data1,data2])
result2 = pd.concat([data3,data4])
result3 = pd.concat([result,result2])
```

```python
result3.to_csv("부산대 놀거리.csv", encoding='UTF-8')
```
